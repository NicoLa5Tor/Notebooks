```python
import matplotlib.pyplot as plt

def torta():
    # Datos de ejemplo: porcentajes de uso de 5 softwares
    softwares = ['Software A', 'Software B', 'Software C', 'Software D', 'Software E']
    porcentajes = [20, 25, 15, 10, 30]  # Porcentajes de uso de cada software
    
    # Configuración de colores para la gráfica
    colores = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
    
    # Crear la gráfica de torta
    plt.figure(figsize=(8, 6))  # Tamaño de la figura
    plt.pie(porcentajes, labels=softwares, colors=colores, autopct='%1.1f%%', startangle=140)
    
    # Añadir título
    plt.title('Top software vulnerable')
    
    # Mostrar la gráfica
    plt.axis('equal')  # Hacer que la gráfica sea un círculo en lugar de una elipse
    plt.show()
def barras():
        # Datos de ejemplo (número de CVEs por máquina)
    maquinas = ['Máquina A', 'Máquina B', 'Máquina C']
    cve_count = [35, 28, 20]
    
    # Crear la gráfica de barras
    plt.figure(figsize=(10, 6))
    plt.bar(maquinas, cve_count, color=['gold','yellowgreen', 'lightcoral'])
    
    # Añadir etiquetas y título
    plt.xlabel('Máquinas')
    plt.ylabel('Número de CVEs')
    plt.title('Top 3 Máquinas por Número de CVEs')
    
    # Mostrar la gráfica
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
torta()
barras()
```
